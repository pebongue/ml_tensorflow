```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('ml-bugs.csv')
df
```

```python
def calcEntropy(arr):
    n_total = len(arr)
    entropy = 0
    for item in set(arr):
        p_item = len(arr[arr == item]) / n_total
        entropy -= p_item*np.log2(p_item)
    return entropy
```

```python
all = df['Species'].values
entropy = calcEntropy(all)
```

```python
green = df[df['Color'] == 'Green'].values[:,0]
not_green = df[df['Color'] != 'Green'].values[:,0]
e_green = (calcEntropy(green) + calcEntropy(not_green)) / 2

blue = df[df['Color'] == 'Blue'].values[:,0]
not_blue = df[df['Color'] != 'Blue'].values[:,0]
e_blue = (calcEntropy(blue) + calcEntropy(not_blue)) / 2

brown = df[df['Color'] == 'Brown'].values[:,0]
not_brown = df[df['Color'] != 'Brown'].values[:,0]
e_brown = (calcEntropy(brown) + calcEntropy(not_brown)) / 2

print('Information Gain for green is {}'.format(entropy - e_green))
print('Information Gain for blue is {}'.format(entropy - e_blue))
print('Information Gain for brown is {}'.format(entropy - e_brown))
```

```python
for i in range(10,25):
    arr1 = df[df['Length (mm)'] < i].values[:,0]
    arr2 = df[df['Length (mm)'] >= i].values[:,0]
    e_arr = (calcEntropy(arr1) + calcEntropy(arr2)) / 2
    print('Information Gain for Length < {} is {}'.format(i, entropy - e_arr))
```
